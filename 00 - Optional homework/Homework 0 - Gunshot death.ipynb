```python
%pylab inline
import pandas as pd
```

```python
df = pd.read_csv("GunShot.csv", sep=',', header=0, index_col=0)
df
```

```python
deaths = df.loc[(df['Intent']=='None selected') \
       & (df['Gender']=='None selected') \
       & (df['Age']=='None selected') \
       & (df['Race']=='None selected') \
      ]['Deaths'].values[0]
deaths
```

But nearly two-thirds of gun deaths are suicides.

```python
suicide = df.loc[(df['Intent']=='Suicide') \
       & (df['Gender']=='None selected') \
       & (df['Age']=='None selected') \
       & (df['Race']=='None selected') \
      ]['Deaths'].values[0]
suicide / deaths * 100
```

More than 85% of suicide victims are male...

```python
male_suicide = df.loc[(df['Intent']=='Suicide') \
       & (df['Gender']=='Male') \
       & (df['Age']=='None selected') \
       & (df['Race']=='None selected') \
      ]['Deaths'].values[0]
male_suicide/suicide * 100
```

… and more than half of all suicides are men age 45 or older.

```python
##35 or 45??
male_more_35_deaths = df.loc[(df['Intent']=='Suicide') \
       & (df['Gender']=='Male') \
       & (df['Race']=='None selected') \
       & (df['Age'].isin(['35 - 64', '65+']))
      ]['Deaths'].sum()
male_more_35_deaths / male_suicide * 100
```

Another third of all gun deaths — about 12,000 in total each year — are homicides.

```python
homicide = df.loc[(df['Intent']=='Homicide') \
       & (df['Gender']=='None selected') \
       & (df['Age']=='None selected') \
       & (df['Race']=='None selected') \
      ]['Deaths'].values[0]
print("deaths by homicide: {}".format(homicide))
homicide/deaths * 100
```

```python
#young men = what age?
young_men_homicide = df.loc[(df['Intent']=='Homicide') \
       & (df['Gender']=='Male') \
       & (df['Age'].isin(['Under 15', '15 - 34'])) \
       & (df['Race']=='None selected') \
      ]['Deaths'].sum()
print("young_men_homicide: {}".format(young_men_homicide))
young_men_homicide / homicide * 100
```

```python
black_young_men_homicide = df.loc[(df['Intent']=='Homicide') \
       & (df['Gender']=='Male') \
       & (df['Age'].isin(['Under 15', '15 - 34'])) \
       & (df['Race']=='Black') \
      ]['Deaths'].sum()
black_young_men_homicide / young_men_homicide * 100
```

Women are far less likely to be gun homicide victims — about 1,700 of them are killed each year, many in domestic violence incidents.

```python
women = df.loc[(df['Intent']=='Homicide') \
       & (df['Gender']=='Female') \
       & (df['Age'] == 'None selected') \
       & (df['Race']=='None selected') \
      ]['Deaths'].sum()
women
```

The remaining gun deaths are accidents or are classified as undetermined.

```python
accidents = df.loc[(df['Intent']=='Accident') \
       & (df['Gender']=='None selected') \
       & (df['Age'] == 'None selected') \
       & (df['Race']=='None selected') \
      ]['Deaths'].sum()
unknown = df.loc[(df['Intent']=='Unknown') \
       & (df['Gender']=='None selected') \
       & (df['Age'] == 'None selected') \
       & (df['Race']=='None selected') \
      ]['Deaths'].sum()

print('Accident deaths: {}, unknown deaths: {}'.format(accidents, unknown))
```
